```python
#8
def histogram(items):
    for n in items:
        output=""
        times=n
        while times>0:
            output=output+"*"
            times=times-1
        print(output)
histogram([3,6,4,3,6,12])        
```

```python
#9
def find_longest_word(words_list):
    words_len=[]
    for n in words_list:
        words_len.append((len(n),n))
        words_len.sort()
    return words_len [-1][1]
words=input("enter sevralwords:").split()
print(find_longest_word(words))
```
